```python
#import libraries
import pandas as pd
```

```python
#loading IMDB movie_metadata.csv
movie_df = pd.read_csv('movie_metadata.csv')
```

```python
movie_df.head()
```

```python
movie_df.info()
```

```python
#Creating dataframe of key columns from original df
movie_df_keyc = movie_df[['actor_1_name','actor_2_name','actor_3_name','genres','director_name','budget','gross']].copy()
```

```python
movie_df_keyc.head()
```

```python
#function to calculate profitability
def calc_profit(row):
    return (row['gross']-row['budget'])
```

```python
#Adding new column profitability 
movie_df_keyc['profitability']=movie_df_keyc.apply(calc_profit,axis=1)
```

```python
movie_df_keyc.head()
```

```python
#sorting dataframe by profitability in decreasing order 
sort_by_profit = movie_df_keyc.sort_values(['profitability'], ascending=[False])
```

```python
#Computing the top 10 genres in decreasing order by their profitability
sort_by_profit[['genres','profitability']].head(10)
```

```python
#Returning the top 10 actors or directors in decreasing order by their profitability
sort_by_profit[['actor_1_name','actor_2_name','actor_3_name']].head(10)
```

```python
#Finding the best actor, director pairs (up to 10) that have the highest IMDB_ratings, if there are indeed any such pairs.
def best_actor_director(row):
    #Considering imdb rating more than or equal to 8 as hightest IMDB_rated
    if row['imdb_score']>=8:
        return (row['actor_1_name'],row['director_name'])
```

```python
movie_df.apply(best_actor_director,axis=1).head(10)
```
